## Univariate Exploration

```python
def univariiable_bar(columnsName, titleName, xaxisName, yaxisName, hr=False, ordered=False):
    if hr == False:
        plt.figure(figsize=[10, 5])
        base_color = sb.color_palette()[0]
        orders = list(prosper_loan_copy[columnsName].value_counts().index)
        
        if ordered == False:
            sb.countplot(data=prosper_loan_copy, x=columnsName, color=base_color)
        else: 
            sb.countplot(data=prosper_loan_copy, x=columnsName, color=base_color, order=orders)
        plt.title(titleName, fontsize=15)
        plt.xlabel(xaxisName, fontsize=12)
        plt.ylabel(yaxisName, fontsize=12)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12);
        
    else:
        plt.figure(figsize=[10, 5])
        base_color = sb.color_palette()[0]
        orders = list(prosper_loan_copy[columnsName].value_counts().index)
        
        if ordered == False:
            sb.countplot(data=prosper_loan_copy, y=columnsName, color=base_color)
        else: 
            sb.countplot(data=prosper_loan_copy, y=columnsName, color=base_color, order=orders)
        
        plt.title(titleName, fontsize=15)
        plt.xlabel(xaxisName, fontsize=12)
        plt.ylabel(yaxisName, fontsize=12)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12);
```

```python
def bar_dtributioin(columnsName_x, columnsName_y, graphetile):
    order_cat = prosper_loan_copy.groupby([columnsName_x]).mean()[columnsName_y].sort_values(ascending=False).index
    plt.figure(figsize=[15, 5])
    base_color = sb.color_palette()[0]
    sb.barplot(data=prosper_loan_copy, x=columnsName_x, y=columnsName_y, color=base_color, order=order_cat, 
               errwidth=0)
    plt.title(graphetile, fontsize=15)
    plt.xlabel('Occupation Categories', fontsize=13)
    plt.ylabel('Income Distribution', fontsize=13)
    plt.xticks(rotation=90);
```
